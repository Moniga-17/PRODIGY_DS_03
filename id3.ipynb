```python
# load an example dataset
from vega_datasets import data
cars = data.cars()

import altair as alt

interval = alt.selection_interval()

points = alt.Chart(cars).mark_point().encode(
  x='Horsepower',
  y='Miles_per_Gallon',
  color=alt.condition(interval, 'Origin', alt.value('lightgray'))
).properties(
  selection=interval
)

histogram = alt.Chart(cars).mark_bar().encode(
  x='count()',
  y='Origin',
  color='Origin'
).transform_filter(interval)

points & histogram
```

```python
from sklearn import tree
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_text
from sklearn.ensemble import RandomForestClassifier
from matplotlib import pyplot as plt
iris=load_iris()
dtree=DecisionTreeClassifier(random_state=0,max_depth=5)
dtree.fit(iris.data,iris.target)
r=export_text(dtree,feature_names = iris['feature_names'])
print(r)
```

```python
py=plt.figure(figsize=(10,15))
tree.plot_tree(dtree)
```
